```python
import pybaobabdt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pygraphviz 
from sklearn import tree

import warnings
warnings.filterwarnings("ignore") 
```

## Read the dataset

```python
df = pd.read_csv('penguins_classification.csv')
df.head()
```

```python
df['Species'] = df['Species'].str.encode('utf-8') 
```

## Visualizing using Pybaobabdt

```python
from sklearn.tree import DecisionTreeClassifier
y = list(df['Species'])
features = list(df.columns)
features.remove('Species')
X = df.loc[:, features]

clf = DecisionTreeClassifier().fit(X,y)
```

```python

ax = pybaobabdt.drawTree(clf, size=10, dpi=72, features=features, colormap='Spectral')
```

```python
ax.get_figure().savefig('tree_example.svg', format='png', dpi=300, transparent=True)
```

## Customizations

### Colormaps

```python
colormaps = ['plasma', 'magma', 'viridis', 'tab20']
for color in colormaps:
    print(f'Colormap : {color}')
    ax = pybaobabdt.drawTree(clf, size=10, dpi=72, features=features, colormap=color)
    
```

#### Defining a custom colormap

```python
from matplotlib.colors import ListedColormap
colors = ["green", "gray", "gray"]
colorMap = ListedColormap(colors)
ax = pybaobabdt.drawTree(clf, size=10, features=features, ratio=0.8,colormap=colorMap)
```

### Ratio

```python
print(f'Ratio  = 0.5')
ax = pybaobabdt.drawTree(clf, size=10, dpi=300, features=features, ratio=0.5,colormap='viridis')
```

```python
print(f'Ratio  = 1')
ax = pybaobabdt.drawTree(clf, size=10, dpi=300, features=features, ratio=1,colormap='viridis')
```

### Saving the image


```python
ax.get_figure().savefig('claasifier_tree.png', format='png', dpi=300, transparent=True)
```
